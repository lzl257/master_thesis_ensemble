```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(2020)
x = np.random.normal(0, 1, 100)
eps = np.random.normal(0, 0.3, 100)
y = 0.5 * x + eps


fig, ax = plt.subplots(figsize=(8, 8))
ax.scatter(x, y)
ax.scatter([2], [1], s=60, color='r')
ax.scatter([-1.5], [0.8], s=60, color='r')
ax.annotate('A', (-1.4, 0.8), fontsize='xx-large')
ax.annotate('B', (1.8, 1), fontsize='xx-large')
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')

fig.savefig('output/figures/figure232.png')
```
